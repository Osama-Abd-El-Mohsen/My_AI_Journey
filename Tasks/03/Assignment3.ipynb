# Activity 1

1. Define T (Task) : This is a **Classification** problem.
The bot must decide whether to **NOT SELL(N) `(hold/buy/ignore)`, Sell(P)**.

2. Define P (Performance Measure) :
    
   AI Suggested `(Net Profit/Loss (P&L) or Return on Investment (ROI))` will suit this case but for Precision or Recall

   **Precision** is more important than **recall** for a trading bot because minimizing false trades (financial losses) is more critical than capturing every profitable opportunity.
   


3. Define E (Experience)


    - Time series of price (open, high, low, close) at very small intervals.

    - Volume and order book features (bid/ask prices, depth).

    - Technical indicators (moving averages, RSI, MACD, volatility).

    - Possibly news sentiment / events around trades.

    - Labels: at each time, whether a Sell there would have been profitable according to some rule (e.g., “price drops at least X% after this point”).

4. Critical Thinking Question : 

   - the bot assumed that the statistical properties (mean, variance, dependencies) of the market data (E) learned from 2010-2020 would remain constant over time. This is violated in 2024 because market regimes, regulatory environments, and trading dynamics change (e.g., changes in central bank policy, new trading algorithms, shifts in liquidity). This concept is known as data drift or concept drift in ML.

# Activity 2

Total Production: 1,000,000 chips.

Defective Chips: 500.

Good Chips: 999,500.





**Target** : Detect Defective Chips **Good (N) / Defective (P)**

`accuracy = TP + TN / TP + TN + FP + FN` = 99.95 % 

But that is  a  missleading duo to (Imbalanced Data)
Defective chips are very rare: only 500 out of 1,000,000.
The model never detects any defective chip, but still gets 99.95% accuracy because almost all chips are good.

In reality, this model is useless for the business:
It misses 100% of the defective chips, which is exactly what we care about detecting.

`precesion = TP / TP + FP` = 0 %

`recall = TP / TP + FN` = 0 % 

recall is a better metric because 

**(FN) ->  Chip is actually Defective, but model predicts Good** 

Result: Bad chip goes to customer → **product failure, safety risks, recalls, reputation damage, legal costs.**

**(FP): Chip is actually Good, but model predicts Defective.** 

Result: You scrap or re-test a good chip → **some extra cost, but no customer harm**.
