## 2. Machine Learning Paradigms
**Exercise 3:** Describe Supervised, Unsupervised, Semi-supervised, Reinforcement Learning.

**Supervised Learning** : Supervised learning learns from labeled data Ex : Classification (Spam Filters) - Regression (House Price)

**Unsupervised Learning** : Unsupervised learning works with unlabeled data The goal is to find hidden patterns and structure in the data. : Clustering (Group similar items (Customer segmentation) ) - Dimensionality Reduction (Simplify data)

**Semi-Supervised Learning** : Semi-supervised learning combines: A small amount of labeled data - A large amount of unlabeled data


**Reinforcement Learning** : Reinforcement learning trains an agent by letting it: - Interact with an environment - Take actions - Receive rewards or penalties

## 3. Regression & Classification
**Exercise 5:** Compare Linear vs Polynomial Regression.

| Feature | Linear Regression | Polynomial Regression |
|--------|-------------------|--------------------------|
| Type | Linear model | Non-linear model (still linear in parameters) |
| Relationship shape | Straight line | Curved line |
| Formula | y = mx + b | y = a₀ + a₁x + a₂x² + ... + aₙxⁿ |
| Complexity | Simple | More complex (depends on polynomial degree) |
| Overfitting risk | Low | Higher with large degree |
| Interpretability | Easy to interpret | Harder to interpret |
| Data required | Works with small datasets | Needs more data for high-degree polynomials |
| Best for | Data with linear trend | Data with curved trend |
| Example use | Sales vs advertising spend | Population growth patterns |
| Visualization | Straight best-fit line | Smooth curve fitting |

## 4. Evaluation Metrics
**Exercise 8:** Explain MSE, RMSE, MAE.

| Metric | Definition | Formula | Sensitivity to Outliers | Interpretability |
|--------|-------------|----------|---------------------------|-------------------|
| **MSE** | The average of the squared differences between actual values and predicted values. | avg((y − ŷ)²) | High | Low (units are squared) |
| **RMSE** | The square root of MSE, representing the typical prediction error in original units. | √MSE | High | High (original units) |
| **MAE** | The average of the absolute differences between actual values and predicted values. | avg(\|y − ŷ\|) | Low | High (original units) |

| **Precision** | TP / (TP + FP) |
| **Recall (Sensitivity)** | TP / (TP + FN) |
| **F1-Score** | 2 × (Precision × Recall) / (Precision + Recall) |

#  Logistic Regression

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression

modle = LogisticRegression(max_iter=1000)
iris  = load_iris(as_frame=True)
```

```python
iris["data"].head()
```

```python
iris["data"].shape
```

```python
X = iris["data"].to_numpy()
y = iris["target"].to_numpy()
```

## 3 Classes

```python
np.unique(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,      
    random_state=42,    
    stratify=y          
)
```

```python
modle.fit(X_train,y_train)
```

```python
y_pred = modle.predict(X_test)
```

```python
acc  = accuracy_score(y_test, y_pred)
print(f"Accuracy : {acc:.4f}")
```

```python
print("\nClassification report:\n", classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python
con  = confusion_matrix(y_test, y_pred)
sns.heatmap(con, annot=True)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
tree_modle = DecisionTreeClassifier()
```

```python
tree_modle.fit(X_train,y_train)
```

```python
y_pred = tree_modle.predict(X_test)
```

```python
from sklearn.tree import export_text

rules = export_text(
    tree_modle,
    feature_names=list(["sepal length","sepal width","petal length","petal width"])  
)

print(rules)
```

```python
class_names = iris.target_names

def plot_decision_boundary(ax, model, X, y, title, class_names):
    X = np.asarray(X)
    y = np.asarray(y)

    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300),
        np.linspace(y_min, y_max, 300),
    )

    grid = np.c_[xx.ravel(), yy.ravel()]   
    Z = model.predict(grid)                
    Z = Z.reshape(xx.shape)

    cs = ax.contourf(xx, yy, Z, alpha=0.3)

    
    for idx, name in enumerate(class_names):
        ax.scatter(
            X[y == idx, 0],
            X[y == idx, 1],
            label=name,
            edgecolor="k",
            s=40,
        )

    ax.set_xlabel("Petal length (cm)")
    ax.set_ylabel("Petal width (cm)")
    ax.set_title(title)
    ax.legend()


```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
plot_decision_boundary(
    axes[0],
    modle,
    X_2d, y,
    "Logistic Regression (petal length vs petal width)",
    iris.target_names
)
plot_decision_boundary(
    axes[1],
    tree_modle,
    X_2d, y,
    "Decision Tree  (petal length vs petal width)",
    iris.target_names
)
plt.tight_layout()
plt.show()
```

**Exercise 13:** Differences between Hard vs Soft Clustering.

| Feature                  | Hard Clustering                                 | Soft Clustering                                                     |
| ------------------------ | ----------------------------------------------- | ------------------------------------------------------------------- |
| Assignment Type          | Each data point belongs to **only one cluster** | Each data point belongs to **multiple clusters with probabilities** |
| Output                   | Single label per data point                     | Probability distribution across clusters                            |
| Overlapping Clusters     | ❌ Not allowed                                   | ✅ Allowed                                                           |
| Uncertainty Modeling     | ❌ No                                            | ✅ Yes                                                               |
| Interpretability         | ✅ Easy to interpret                             | ⚠ Medium, more complex                                              |
| Computational Complexity | ✅ Lower                                         | ❌ Higher                                                            |
| Speed                    | ✅ Faster                                        | ⚠ Slower                                                            |
| Common Algorithms        | K-Means, K-Medoids, Hierarchical                | Gaussian Mixture Models (GMM), Fuzzy C-Means                        |
| Best for                 | Well-separated clusters with low ambiguity      | Overlapping clusters or uncertain group membership                  |
| Example Output           | `Cluster = 2`                                   | Cluster 1 → 20% - Cluster 2 → 60% - Cluster 3 →20%  `[0.2, 0.6, 0.2]`                                                 |
